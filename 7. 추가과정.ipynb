# 사전 정확도 검증 - 1차

```python
import pandas as pd
# 사전 불러오기
aa = pd.read_csv('./total_label_re_2.csv', index_col='ngrams')
aa
```

```python
# Training Set 불러오기
bb = pd.read_csv('./test_set_final.csv')
bb
```

```python
import math

# 예측 및 모델 검증
def predict(test_set, dic_df, sum_1, sum_0) :
#     dic_df["polarityscore"] = dic_df["매파prob"]/dic_df["비둘기파prob"]
#     dic_df["intensity"] = [x if x > 1 else 1/x for x in dic_df['polarityscore']]
#     dic_df = dic_df.loc[dic_df.intensity>intensity]

    dic = dic_df.to_dict('index')
    
    prob_pos_tot = math.log(sum_1/(sum_1+sum_0))
    prob_neg_tot = math.log(sum_0/(sum_1+sum_0))
    
    TP = FN = FP = TN = cnt = 0
    predict=[]

    for i, row in test_set.iterrows():
        ngrams = row['ngrams and tokens'].split(',')
        label = row['label']
        prob_predict = prob_pos = prob_neg = 0

        tmp_token = ""
        tmp_pos = 0
        tmp_neg = 0
        try :
            for token in ngrams :
                if token == '' :
                    continue 
                
                if token in dic.keys() :
                    prob_pos += math.log(dic[token]['매파prob'])
                    prob_neg += math.log(dic[token]['비둘기파prob'])
                    
            prob_pos = math.exp(prob_pos_tot + prob_pos)
            prob_neg = math.exp(prob_neg_tot + prob_neg)
        except Exception as e :
            print("===============")
                        
            print(str(e))
            print("prob_pos_tot : {:10.4f}".format(prob_pos_tot))
            print("prob_pos : {:10.4f}".format(prob_pos))
            print("prob_neg_tot : {:10.4f}".format(prob_neg_tot))
            print("prob_neg : {:10.4f}".format(prob_neg))
            print("sum_1 : {:10.4f}".format(sum_1))
            print("sum_0 : {:10.4f}".format(sum_0))
            print(ngrams)


        if prob_pos+prob_neg == 0 :
            prob_predict = 0
        else :
            prob_predict = prob_pos/(prob_pos+prob_neg)

        if prob_predict > 0.5 and label == 1.0 :
            predict.append(1.0)
            TP += 1
        elif prob_predict > 0.5 and label == -1.0 :
            predict.append(1.0)
            FP += 1
        elif prob_predict < 0.5 and label == 1.0 :
            predict.append(-1.0)
            FN += 1
        elif prob_predict < 0.5 and label == -1.0 :
            predict.append(-1.0)
            TN += 1
            
    print('TP={} / TN={} / FP={} / FN={}'.format(TP, TN, FP, FN))
    #ACC = (TP + TN)  / (전체 데이타 수 = P + N)
    ACC = (TP + TN) / (TP + TN + FP + FN) * 100
    # ERR = (FN+FP) / (전체 데이타수 = P+N)
    ERR = (FN+FP) / (TP + TN + FP + FN) * 100
    # SN = (TP) / P
    SN = (TP) / (TP + FN) * 100
    # PREC = TP / (TP+FP)
    PREC = TP / (TP+FP) * 100
    print("ACC={:7.3f}% / ERR = {:7.3f}% / SN = {:7.3f}% / PREC = {:7.3f}%".format(ACC, ERR, SN, PREC))
```

```python
# 사전에서의 매파 / 비둘기파 각각 총 빈도 수 
sum_1 = 1793980
sum_0 = 1995406
```

```python
# 힘수 대입해서 정확도 측정
predict(bb,aa, sum_1, sum_0)
```
