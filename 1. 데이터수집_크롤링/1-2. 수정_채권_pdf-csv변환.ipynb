```python
from tika import parser
from glob import glob
import pandas as pd
from os import rename, listdir
```

```python
newnew = glob('*.pdf')
```

```python
newnew
```

```python
df = pd.DataFrame(columns=['date','com','script'])
    
for title in newnew:
    filename = title
    #title = title.replace('\\','/')
    try:
        raw = parser.from_file(title)
    except:
        rename(title, "1.pdf")
        raw = parser.from_file('1.pdf')
        rename("1.pdf", title)
        
    secname  = filename[8:].split('.')[0]
    datestr = '20' + filename[:8]
    try:
        df = df.append(pd.DataFrame([[ datestr,secname,raw['content'].replace("\n","")]], columns=['date','com','script']))
    except:
        print(title)
```

```python
df.to_csv('bondreport2.csv',encoding='UTF-8')
```
