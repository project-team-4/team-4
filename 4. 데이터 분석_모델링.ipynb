# 의사록 라벨링(trainingset_minutes)

```python
import pandas as pd


df = pd.read_csv('./전처리/re_tokenization_ngrams.csv', encoding=u'utf-8')
callrate = pd.read_csv('./callrate.csv',encoding='utf-8')
callrate = callrate[['date', 'label']]
callrate['date'] = callrate['date'].str.replace(" ", "")

training_set = df[['date', 'tokens2', 'ngrams2', 'tokens3', 'ngrams3']]

merge = pd.merge(training_set, callrate, on = 'date', how = 'left')

#NAN값을 '0'으로 채우기
merge['tokens2']=merge['tokens2'].fillna("0")
merge['tokens3']=merge['tokens3'].fillna("0")
merge['ngrams2']=merge['ngrams2'].fillna("0")
merge['ngrams3']=merge['ngrams3'].fillna("0")

#ngrams와 tokens 합칠 때 사이에 콤마(,) 넣기
merge['ngrams and tokens'] = merge['ngrams2'] + ',' + merge['tokens2'] + ',' + merge['ngrams3'] + ',' + merge['tokens3']


total = merge.loc[merge['ngrams and tokens']!= '0,0,0,0' ]               
total = total.loc[total['label']!= 0 ]        
df_total = pd.DataFrame(total)
df_total

for i in range(len(total)):
    try:
        b1 = total['ngrams and tokens'][i]
        b1= b1.replace("0,", "")
        b1= b1.replace(",0", "")
        df_total['ngrams and tokens'][i]= b1
    except:
        i=i+1


# total.to_csv('merge_label_minutes(190527).csv', header=True, index=False)  
pd.read_csv('./라벨링_의사록/merge_label_minutes(190527).csv')
```

# 뉴스 라벨링(trainingset_news)_2005~2017

```python
#2005년 뉴스기사 라벨링, 2006-2017년도 이와 같은 방법으로 실행
import pandas as pd

df_news = pd.read_csv('./전처리/news_preprocessed/news_2005_preprocessed.csv',encoding='utf-8')
callrate = pd.read_csv("./callrate.csv",encoding='utf-8')
callrate = callrate[['date', 'label']]
callrate['date'] = callrate['date'].str.replace(" ", "")

training_set = df_news[['date', 'tokens', 'ngram']]

merge = pd.merge(training_set, callrate, on = 'date', how = 'left')

#NAN값을 '0'으로 채우기
merge['tokens']=merge['tokens'].fillna("0")
merge['ngram']=merge['ngram'].fillna("0")

#ngram과 token 더해 'ngrams and tokens' 값 지정 
if(str(merge['ngram'])== "0"):
    merge['ngrams and tokens'] = merge['tokens']
elif(str(merge['tokens'])== "0"):
    merge['ngrams and tokens'] = merge['ngram']
else:
    merge['ngrams and tokens'] = merge['ngram'] + "," + merge['tokens']

for i in range(len(merge)):
    b = merge['ngrams and tokens'][i]
    b= b.replace("0,", "")
    b= b.replace(",0", "")
    merge['ngrams and tokens'][i]=b
    

list_ = [-1.0, 0, 1.0] 
for i in reversed(range(len(merge))):
    k = merge['label'][i]
    if k not in list_:
        k = merge['label'][i+1]
        merge['label'][i] = k

#라벨이 0이 아닌 경우만 저장
total = merge.loc[merge['label']!= 0 ]

# total.to_csv('./라벨링_뉴스기사/2005_merge_label.csv', header=True, index=False)
pd.read_csv('./라벨링_뉴스기사/2005_merge_label.csv')
```

# 각 ngrams에 대한 매파, 비둘기파 count (딕셔너리 생성)

```python
from collections import defaultdict
import pandas as pd 
import numpy as np
import math
import os
from os import listdir

file_url = "./merge_label/"
files = listdir(file_url)

doccnt1 = 0
doccnt0 = 0
wordfreq = defaultdict(lambda : [0, 0])

#파일들을 차례로 불러옴
for filenames in files:    
    a = pd.read_csv(file_url + filenames)
    
    #','을 기준으로 split
    for nt, row in a.iterrows():
        words = row['ngrams and tokens'].split(',')

        for word in words:
            if row['label'] == 1.0:
                wordfreq[word][1] += 1
            else:
                wordfreq[word][0] += 1



a = pd.DataFrame.from_dict(wordfreq, orient='index')

# a.to_csv("./merge_lable_dict_2.csv")

pd.read_csv("./merge_lable_dict_2.csv")
```

# 0을 비둘기파, 1을 매파로 rename하여 저장

```python
label_count = pd.read_csv("./merge_lable_dict_2.csv")
label_count = label_count.rename(columns = {'Unnamed: 0': 'ngrams'})
label_count = label_count.rename(columns = {'0': '비둘기파'})
label_count = label_count.rename(columns = {'1': '매파'})
label_count.to_csv("./merge_lable_dict_2.csv",  index = False)
```

# 비둘기파와 매파의 prob구하기

```python
k =0.5
doccnt1 = 0
doccnt0 = 0

for key, (cnt0, cnt1) in wordfreq.items() :
    if cnt1 > 0 :
        doccnt1 += cnt1
    if cnt0 > 0 :
        doccnt0 += cnt0
        
wordprobs = defaultdict(lambda : [0,0])

for key, (cnt0, cnt1) in wordfreq.items() :
    wordprobs[key][1] = (cnt1 + k) / (doccnt1 + 2*k)   
    wordprobs[key][0] = (cnt0 + k) / (doccnt0 + 2*k)

b = pd.DataFrame.from_dict(wordprobs, orient='index')

b.to_csv("./merge_lable_dict_prob_2.csv")
b
```
