```python
import pandas as pd
from ekonlpy.sentiment import MPCK
mpck = MPCK()
df = pd.read_csv('./minutes_section_1.csv', encoding=u'utf-8')
```

```python
df
```

```python


df["tokens2"] = ''
df["tokens3"] = ''
df["ngrams2"] = ''
df["ngrams3"] = ''


for i, row in df.iterrows():
    try :
        text2 = row['section2']
        text3 = row['section3']

        tokens2 = mpck.tokenize(text2)
        df['tokens2'][i] = ','.join(tokens2)
        print(tokens2)
        
        ngrams2 = mpck.ngramize(tokens2)
        df['ngrams2'][i] = ','.join(ngrams2)
        print(ngrams2)
        
        tokens3 = mpck.tokenize(text3)
        df['tokens3'][i] = ','.join(tokens3)
        print(tokens3)
        
        ngrams3 = mpck.ngramize(tokens3)
        df['ngrams3'][i] = ','.join(ngrams3)
        print(ngrams3)
        
        print(str(i) + "번=============")
    except Exception as ex :
        print(str(i) + "번=============" + str(ex))
df.to_csv('minutes_section_sum.csv')
```
