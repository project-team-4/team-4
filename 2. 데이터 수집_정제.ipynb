# 1-2-3. 2017년 뉴스기사 필요없는 내용 제거

```python
import pandas as pd
import re

# 2017년 1월 크롤링한 기사 가져오기 (같은 방법으로 1월부터 12월까지 실행)
text = pd.read_csv('./@@prepro_201701.csv')

# 중복 제거 후 'preprocessed' column 생성
text = text.drop_duplicates(['script'])
text['preprocessed'] = 6


for i in range(len(text)):
    
    txt = text['script'][i]

    
# 내용에서 float(실수)가 아닌 부분에 대해 다음과 같은 작업
    if type(txt) != float:

        
        email_loc = re.search('[a-zA-Z0-9+-_.]+@yna.co.kr', txt)
        if email_loc != None :
            txt = txt[:email_loc.start()].strip()

        email_loc2 = re.search('[a-zA-Z0-9+-_.]+@edaily.co.kr', txt)
        if email_loc2 != None :
            txt = txt[:email_loc2.start()].strip()

        email_loc3 = re.search('[a-zA-Z0-9+-_.]+@yonhapnews.co.kr', txt)
        if email_loc3 != None :
            txt = txt[:email_loc3.start()].strip()

        author_loc = re.search('기자 =', txt)
        if author_loc != None :
            txt = txt[author_loc.start()+4:].strip()

        tag_loc = re.search('\[\기업은행', txt)
        if tag_loc != None :
            txt = txt[:tag_loc.start()].strip()

        copyright_loc = re.search('Copyright', txt)
        if copyright_loc != None :
            txt = txt[:copyright_loc.start()].strip() #strip()은 공백제거

        txt = txt.replace("(서울=연합뉴스)","")
        txt = txt.replace('(뉴욕=연합인포맥스) 이종혁 특파원',"")
        txt = txt.replace('(뉴욕=연합인포맥스) 신은실 특파원',"") 
        txt = re.sub('\[+[0-9a-zA-Z가-힣+./=+ -?]+\]','',txt).strip()
        txt = re.sub('\([0-9a-zA-Z가-힣+./=+ -©]+맥스\]','',txt).strip()
        txt = re.sub('([0-9]{4}).([0-9]).([0-9])','',txt).strip()
        txt = re.sub('[●▲▶■◀◆☞◇△]','',txt).strip()

        sub1 = "(★별도 표 있음)"
        txt = txt.replace(sub1, '')

        print("******기사"+ str(i+1) + "번*******")
        print(txt)
        print("========")
        text['preprocessed'][i] = txt

# 공백 있는 행을 drop
text = text.dropna()

text.to_csv('./news_201701_preprocessed.csv', header=True, index=False)
```

```python
# https://colab.research.google.com/drive/1lOwqZtUZCj26LghriMvg_aX5qvCjbT7K#scrollTo=iHKYE5vlhxE1의 코드 사용

def tidy_sentences(section):
    #print("tidy_sentences=====================")
    #print(section)
    sentence_enders = re.compile(r'((?<=[함음됨임봄짐움])(\s*\n|\.|;)|(?<=다)\.)\s*')
    #print(sentence_enders)
    splits = list((m.start(), m.end()) for m in re.finditer(sentence_enders, section))
    #print(splits)
    starts = [0] + [i[1] for i in splits]
    #print(starts)
    ends = [i[0] for i in splits]
    #print(ends)
    sentences = [section[start:end] for start, end in zip(starts[:-1], ends)]
    #print(sentences)
    for i, s in enumerate(sentences):
        sentences[i] = (s.replace('\n', ' ').replace(' ', ' ')) + '.'

    text = '\n'.join(sentences) if len(sentences) > 0 else ''
    return sentences, text
```

```python
# https://colab.research.google.com/drive/1lOwqZtUZCj26LghriMvg_aX5qvCjbT7K#scrollTo=iHKYE5vlhxE1의 코드 사용

def preprocess_minutes(minutes):
    pos = re.search('(.?국내외\s?경제\s?동향.?과 관련하여,?|\(가\).+경제전망.*|\(가\) 국내외 경제동향 및 평가)\n?\s*일부 위원은', minutes, re.MULTILINE)
    s1 = pos.start() if pos else -1
    pos = re.search('(.?외환.?국제금융\s?동향.?과 관련하여.*|\(나\) 외환.국제금융\s?(및 금융시장)?\s?동향)\n?\s*(일부 위원은|대부분의 위원들은)', minutes, re.MULTILINE)
    s2 = pos.start() if pos else -1
    pos = re.search('(.?금융시장\s?동향.?과 관련하여,?|\(다\) 금융시장\s?동향)\n?\s*일부 위원은', minutes, re.MULTILINE)
    s3 = pos.start() if pos else -1
    pos = re.search('((\((다|라)\) )?.?통화정책\s?방향.?에 관한 토론,?|이상과 같은 의견\s?교환을 바탕으로.*통화정책\s?방향.*에.*토론.*)\n?', minutes,
                    re.MULTILINE)
    s4 = pos.start() if pos else -1
    pos = re.search('(\(4\) 정부측 열석자 발언.*)\n?', minutes, re.MULTILINE)
    s5 = pos.start() if pos else -1
    pos = re.search('(\(.*\) 한국은행 기준금리 결정에 관한 위원별 의견\s?개진|이상과 같은 토론에 이어 .* 관한 위원별 의견개진이 있었음.*)\n?', minutes,
                    re.MULTILINE)
    s6 = pos.start() if pos else -1
    # pos = re.search('(\(\s?.*\s?\) ()(심의결과|토의결론))\n?', minutes, re.MULTILINE)
    # s7 = pos.start() if pos else -1
    positer = re.finditer('(\(\s?.*\s?\) ()(심의결과|토의결론))\n?', minutes, re.MULTILINE)
    s7 = [pos.start() for pos in positer if pos.start() > s6]
    s7 = s7[0] if s7 else -1


    # 국내외 경제동향
    bos = s1
    eos = s2
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('(일부|대부분의) 위원들?은', section, re.MULTILINE)
    bos = pos.start() if pos else -1
    section = section[bos:] if bos >= 0 else section
    section1, section1_txt = tidy_sentences(section)
    #print(section)
    #print("=======")
    #print(section1)
    
    # 외환․국제금융 동향
    bos = s2
    eos = s3 if s3 >= 0 else s4
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('(일부|대부분의) 위원들?은', section, re.MULTILINE)
    bos = pos.start() if pos else -1
    section = section[bos:] if bos >= 0 else section
    section2, section2_txt = tidy_sentences(section)
    #print(section)

    # 금융시장 동향
    bos = s3
    eos = s4
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('(일부|대부분의) 위원들?은', section, re.MULTILINE)
    bos = pos.start() if pos else -1
    section = section[bos:] if bos >= 0 else section
    section3, section3_txt = tidy_sentences(section)

    # 통화정책방향
    bos = s4
    eos = s5 if s5 >= 0 else s6 if s6 >= 0 else s7
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('(일부|대부분의) 위원들?은', section, re.MULTILINE)
    bos = pos.start() if pos else -1
    section = section[bos:] if bos >= 0 else section
    section4, section4_txt = tidy_sentences(section)

    # 위원별 의견 개진
    bos = s6
    eos = s7
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('(일부|대부분의) 위원들?은', section, re.MULTILINE)
    bos = pos.start() if pos else -1
    section = section[bos:] if bos >= 0 else section
    section5, section5_txt = tidy_sentences(section)

    # 정부측 열석자 발언
    bos = s5
    eos = s6
    section = minutes[bos:eos] if bos >= 0 or eos >= 0 else ''
    pos = re.search('정부측 열석자 발언', section, re.MULTILINE)
    bos = pos.end() + 1 if pos else -1
    section = section[bos:] if bos >= 0 else section
    section6, section6_txt = tidy_sentences(section)

    sections = ['Economic Situation', 'Foreign Currency', 'Financial Markets',
                'Monetary Policy', 'Participants’ Views', 'Government’s View']
    section_texts = (section1, section2, section3, section4, section5, section6)
    
    return section_texts

    doc = []

    if any(section_texts):
        for s, (section, sentences) in enumerate(zip(sections, section_texts)):
            for p, text in enumerate(sentences):
                row = (filename, mdate, rdate, section, p, text)
                doc.append(row)
    else:
        print('Empty!')

    return doc
```

```python
import os
from os import listdir
from datetime import timedelta, date
import glob
import re

import pandas as pd

section1 = []
section2 = []
section3 = []
section4 = []
section5 = []
section6 = []
count1 = []
count2 = []
count3 = []
count4 = []
count5 = []
count6 = []


# 행 하나씩 반복 실행
for i, row in df.iterrows() :

# 행의 script 부분에서 작업, 위의 preprocess_minutes 함수 사용
    rr = preprocess_minutes(row['script'])

# '@@@' 기준으로 spilt, 내용 추가하고 개수 측정
    section1.append('@@@'.join(rr[0]))
    section2.append('@@@'.join(rr[1]))
    section3.append('@@@'.join(rr[2]))
    section4.append('@@@'.join(rr[3]))
    section5.append('@@@'.join(rr[4]))
    section6.append('@@@'.join(rr[5]))
    count1.append(len(rr[0]))
    count2.append(len(rr[1]))
    count3.append(len(rr[2]))
    count4.append(len(rr[3]))
    count5.append(len(rr[4]))
    count6.append(len(rr[5]))

    raw_data = {'section1': section1,'section2': section2, 'section3': section3, 'section4': section4, 'section5': section5, 'section6': section6,'count1': count1,'count2': count2,'count3': count3,'count4': count4,'count5': count5,'count6': count6}
    
    dff = pd.DataFrame(raw_data)
    
```
