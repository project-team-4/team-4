```python
import pandas as pd
from ekonlpy.sentiment import MPCK
mpck = MPCK()


df = pd.read_csv('../news/news_2006_preprocessed.csv').dropna()

df["tokens"] = ''
df["ngrams"] = ''

for i, row in df.iterrows():
    try :
        text = row['news']

        tokens = mpck.tokenize(text)
        ngrams = mpck.ngramize(tokens)

        row['tokens'] = ','.join(tokens)
        row['ngrams'] = ','.join(ngrams)
        
        print(str(i) + "번=============")
    except :
        print(str(i) + "번=============오류!!")
df.to_csv('./newssssssss/news_2006_preprocessed.csv')
```
