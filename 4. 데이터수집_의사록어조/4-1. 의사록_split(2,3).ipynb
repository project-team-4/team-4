```python
import pandas as pd 

df = pd.read_csv('C:/Users/user/Downloads/minutes_section_2and3_total.csv')
df

split_sec=pd.DataFrame(columns = ['date','sec2_sec3_split'])

dd= pd.DataFrame();
count=0
p=0

for i in range(len(df)):
    
    df['section2_3'][i] = df['section2_3'][i].replace(str('[]'), "")
    df['section2_3'][i] = df['section2_3'][i].replace(str(']'), "")
    df['section2_3'][i] = df['section2_3'][i].replace(str('['), "")

    dd = df['section2_3'][i].split(str("@@@"))
    
    dd_len=len(dd)
    count=count+i
    
    for j in range(dd_len):
        split_sec.loc[i+p,'sec2_sec3_split'] = dd[j]
        split_sec.loc[i+p,'date'] =  df['date'][i]
        p=p+1
        count=count+j

split_sec
```
