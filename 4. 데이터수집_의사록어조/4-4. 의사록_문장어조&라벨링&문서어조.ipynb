```python
import pandas as pd
from collections import defaultdict
df = pd.read_csv('./total_label_matching222.csv')
df['tone_s']=''
df
```

```python
for i in range(len(df)):
    key1 = df['매파count'][i]
    key2 = df['비둘기파count'][i]
    
    if key1 != key2:
        
        subs = key1 - key2
        sums = key1 + key2

        df['tone_s'][i] = subs / sums
    else:
        df['tone_s'][i] = 0
    
df
```

```python
df.to_csv('total_label_matching_tone_s222.csv', index=False)
```

```python
import pandas as pd
cccc = pd.read_csv('./total_label_matching_tone_s_label222.csv')
dd = cccc.groupby(['date', 'label']).count().reset_index()
tone_tb = pd.merge(dd[dd.label==1][['date','tone_s']], dd[dd.label==-1][['date','tone_s']], on='date')
tone_tb['tone_i'] = 0
tone_tb['tone_i'] = (tone_tb['tone_s_x']-tone_tb['tone_s_y'])/(tone_tb['tone_s_x']+tone_tb['tone_s_y'])
tone_tb
```
