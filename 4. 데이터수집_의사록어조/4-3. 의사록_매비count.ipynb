```python
import pandas as pd
from collections import defaultdict

n_t_merge = pd.read_csv('./sec2_sec3_sentence_split_tokens_ngrams_merge222.csv')
dicts = pd.read_csv('./total_label_re_2.csv')


# 열 생성
n_t_merge['매파count'] = ""
n_t_merge['비둘기파count'] = ""


# 초기값
a_cnt = 0
b_cnt = 0


# 사전 dictionary로 형식변경
dic = defaultdict(lambda: 0)

for i, row in dicts.iterrows() :
    dic[row["ngrams"]] = row["label"]

    
# 매칭    

for i in range(len(n_t_merge)):
    n_t_dic = n_t_merge['ngrams and tokens'][i].split(',')

    for j in range(len(n_t_dic)):
        n_t = n_t_dic[j]
        
        if dic[n_t] == 1:
            a_cnt = a_cnt + 1
        elif dic[n_t] == -1:
            b_cnt = b_cnt + 1
        else:
            continue
            
            
    n_t_merge['매파count'][i] = a_cnt
    n_t_merge['비둘기파count'][i] = b_cnt
    a_cnt = 0
    b_cnt = 0

    
n_t_merge

n_t_merge.to_csv('total_label_matching222.csv', index = False)
```
