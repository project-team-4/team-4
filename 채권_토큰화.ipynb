```python
from ekonlpy.sentiment import MPCK
mpck = MPCK()

import pandas as pd
```

```python
text = pd.read_csv('./bondreport2.csv')

text['token']=""
text['ngram']=""

for i, row in text.iterrows() :
    try :
        aa = row['script']
        tokens = mpck.tokenize(aa)
        row['token'] = '.'.join(tokens)
        ngrams =  mpck.ngramize(tokens)
        row['ngram'] = '.'.join(ngrams)
        
        if i % 100 == 0 :
            print('{}/{} : 채권분석'.format(i, text.shape[0]))
    except Exception as ex :
        print("{} : {}".format(i, str(ex)))
    
text.to_csv('bondreport3.csv',encoding='UTF-8')
    
    #ngram =  mpck.ngramize(tokens)
```
