```python
import pandas as pd


tone_i = pd.read_csv('C:/Users/user/Downloads/tone_i_re.csv')
kospi = pd.read_csv('C:/Users/user/Downloads/kospi.csv')
cpi = pd.read_csv('C:/Users/user/Downloads/cpi_new.csv')
baserate = pd.read_csv('C:/Users/user/Downloads/baserate_new.csv')
```

```python
# '-'를 '.'로 변경해서 date 양식 통일
cpi['date'] = cpi['date'].str.replace("-", ".")
baserate['date'] = baserate['date'].str.replace("-", ".")
```

```python
# tone_mkt, kospi, cpi, baserate 병합
a = pd.merge(tone_i, kospi, on='date')
b = pd.merge(a, cpi, on='date')
c = pd.merge(b, baserate, on='date')
```

```python
# 필요없는 column 제거 
c = c.drop(['tone_s_x', 'tone_s_y'], axis = 1)
```

```python
# kospi 쉼표(,) 제거 후 실수(float)로 변환
c['kospi'] = c['kospi'].str.replace(",", "")
for i in range(len(c)):
    c['kospi'][i] = float(c['kospi'][i])
```

```python
c.to_csv('C:/Users/user/Downloads/index_total.csv', header = True, index = False)
```
