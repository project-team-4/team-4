```python
import pandas as pd



df = pd.read_csv('C:/Users/user/Downloads/news_preprocessed/news_2005_preprocessed_n.csv',encoding='utf-8')
callrate = pd.read_csv("C:/Users/user/Desktop/callrate.csv",encoding='utf-8')
callrate = callrate[['date', 'label']]
callrate['date'] = callrate['date'].str.replace(" ", "")
# callrate

training_set = df[['date', 'tokens', 'ngram']]
# training_set

merge = pd.merge(training_set, callrate, on = 'date', how = 'left')
# callrate #(콜금리)


merge['tokens']=merge['tokens'].fillna("0")
merge['ngram']=merge['ngram'].fillna("0")
if(str(merge['ngram'])== "0"):
    merge['ngrams and tokens'] = merge['tokens']
elif(str(merge['tokens'])== "0"):
    merge['ngrams and tokens'] = merge['ngram']
else:
    merge['ngrams and tokens'] = merge['ngram'] + "," + merge['tokens']

for i in range(len(merge)):
    b = merge['ngrams and tokens'][i]
    b= b.replace("0,", "")
    b= b.replace(",0", "")
    merge['ngrams and tokens'][i]=b
    

merge


llllllllll = [-1.0, 0, 1.0] 
for i in reversed(range(len(merge))):
    k = merge['label'][i]
    if k not in llllllllll:
        k = merge['label'][i+1]
        merge['label'][i] = k
merge.loc[merge['label']== 1.0 ]

total = merge.loc[merge['label']!= 0 ]
total.loc[merge['label']== 0.0 ]
total

total.to_csv('C:/Users/user/Downloads/merge_label/2005_merge_label.csv', header=True, index=False)

# NaN값은 다음 날 값으로 추가
```
