```python
from collections import defaultdict
import pandas as pd 
import numpy as np
import math
import os
from os import listdir

file_url = "C:/Users/user/Downloads/merge_label/"
files = listdir(file_url)

doccnt1 = 0
doccnt0 = 0
wordfreq = defaultdict(lambda : [0, 0])

for filenames in files:
    
    a = pd.read_csv(file_url + filenames)
       
    for nt, row in a.iterrows():
        words = row['ngrams and tokens'].split(',')

        for word in words:
            if row['label'] == 1.0:
                wordfreq[word][1] += 1
            else:
                wordfreq[word][0] += 1

    #문서하나끝...break
            
# for key, (cnt1, cnt0) in wordfreq.items():
#     if cnt1 > 0 :
#         doccnt1 += 1
#     if cnt0 > 0 :
#         doccnt1 += 1
    
# print(wordfreq)

a = pd.DataFrame.from_dict(wordfreq, orient='index')

# a = a.rename(columns = {'Unnamed: 0': 'ngrams'})
# a = a.rename(columns = {'0': '비둘기파'})
# a = a.rename(columns = {'1': '매파'})

a.to_csv("C:/Users/user/Downloads/merge_label/merge_lable_dict_2.csv")
```

```python
k =0.5
doccnt1 = 0
doccnt0 = 0

for key, (cnt0, cnt1) in wordfreq.items() :
    if cnt1 > 0 :
        doccnt1 += cnt1
    if cnt0 > 0 :
        doccnt0 += cnt0
```

```python
wordprobs = defaultdict(lambda : [0,0])

for key, (cnt0, cnt1) in wordfreq.items() :
    wordprobs[key][1] = (cnt1 + k) / (doccnt1 + 2*k)   
    wordprobs[key][0] = (cnt0 + k) / (doccnt0 + 2*k)

b = pd.DataFrame.from_dict(wordprobs, orient='index')

b.to_csv("C:/Users/user/Downloads/merge_label/merge_lable_dict_prob_2.csv")
```
