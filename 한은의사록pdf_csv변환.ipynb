```python
from tika import parser
from glob import glob
import pandas as pd
from os import rename, listdir
```

```python
money = glob('*.pdf')
```

```python
money
```

```python
df = pd.DataFrame(columns=['date','script'])
    
for title in money:
    filename = title
    #title = title.replace('\\','/')
    try:
        raw = parser.from_file(title)
        
        #print(raw['content'])
        #break
    except:
        rename(title, "1.pdf")
        raw = parser.from_file('1.pdf')
        rename("1.pdf", title)
        
    datestr = filename[3:11]
    
    try:
        df = df.append(pd.DataFrame([[datestr,raw['content'].replace("\n","")]], columns=['date','script']))
    except:
        print(title)
```

```python
df.to_csv('moneminutes2.csv',encoding='UTF-8')
```
